```python
import pandas as pd

zdf = pd.read_csv("./zomato.csv", encoding="latin-1")

zdf.drop(columns=['Locality Verbose', 'Switch to order menu', 'Rating color', 'Rating text'], inplace=True)
zdf.dropna(inplace=True)

zdf.head()
```

```python
countryDF = pd.read_excel("./Data/Country-Code.xlsx")
df = pd.merge(zdf, countryDF, how="left", on='Country Code')

df.head()
```

```python
df["Price range"].value_counts()
```

```python
def priceSort(row):
    if row["Price range"] == 1:
        return "Very Cheap"
    elif row["Price range"] == 2:
        return "Cheap"
    elif row["Price range"] == 3:
        return "Moderate"
    elif row["Price range"] == 4: 
        return "Expensive"
    return "Other"

df['Price range'] = df.apply(priceSort, axis=1)
df
```
